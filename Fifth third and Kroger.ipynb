```python
#importing the required python libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#loading the dataset
df=pd.read_csv("fifththird.csv")
pd.DataFrame(df) #Using the python library panda to convert the data into a dataframe
```

```python
print(df.columns) #exploring the column names
print(df.head(5)) #Displaying the first five records
```

```python
df.dtypes #checking for datatypes
```

```python
df.duplicated().sum() #checking for duplicate values
```

```python
pd.DataFrame(round(df.isnull().sum()/df.shape[0] * 100,3), columns = ["Missing"]) #Looking for missing data just so that there will not be any issues when we work with data comparisions
```

```python
df[['TRANSACTION_TYPE','MERCHANT_NAME']]
```

```python
DF =df[(df == 'KROGER').any(axis=1)]
DF
```

```python
DF.count()
```

Initially there were 261513 records(removing the header row). Meaning that there were 261513 customers using fifth third bank credit or debit card to purchase items. 
After using conditions, I found that 119858 fifth third customers have used their card to make purchases at Kroger.

This leaves 141656 fifth third customers who have used their card to purchase elsewhere. 

```python
#Among the 119858 fifththird + Kroger customers ...

DF.TRANSACTION_TYPE.value_counts().nlargest(10).plot(kind = "barh")
plt.title("Credit vs Debit card @ Kroger")
plt.xlabel("Range")
plt.show()

```

```python
plt.rcParams['figure.figsize'] = 14,7
plt.subplot(1,2,1)
DF.MERCHANT_CITY.value_counts().head().plot(kind = "barh", color = sns.color_palette("hls", 5))
plt.xlabel("Cities")
plt.title("Five cities fifth third customers used their card at Kroger the most")

```

KROGER DATASET ANALYSIS

```python
kg=pd.read_csv("5000_products.csv")
pd.DataFrame(kg)
```

```python
kg1=pd.read_csv("5000_households.csv")
pd.DataFrame(kg1)
```

```python
kg2=pd.read_csv("5000_transactions.csv")
pd.DataFrame(kg2)
```

```python
kroger1 = pd.merge(left=kg1, right=kg2, left_on='HSHD_NUM', right_on='HSHD_NUM')
kroger1.shape
kroger1
```

```python
kroger = pd.merge(left=kroger1, right=kg, left_on='PRODUCT_NUM', right_on='PRODUCT_NUM')
kroger.shape
kroger
```

```python
plt.rcParams['figure.figsize'] = 14,7
plt.subplot(1,2,1)
kroger.COMMODITY.value_counts().head().plot(kind = "barh", color = sns.color_palette("hls", 5))
plt.xlabel("Commodities")
plt.title("Top 5 commodities bought")
```

```python
kroger.BRAND_TY.value_counts().nlargest(10).plot(kind = "barh")
plt.title("National vs private products @ Kroger")
plt.xlabel("Range")
plt.show()
```
